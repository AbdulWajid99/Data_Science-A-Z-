```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#for y values
y = np.random.randn(100) #array of 100 elements
print(y+200) #will get in values points
```

```python
#for x values

x = [x for x in range(len(y))]
print(x)
```

```python
#plotting

plt.plot(x,y,'-')
plt.show()
```

```python
# fill under the line
plt.fill_between(x,y,-3,where=(y>-3),facecolor= 'b',alpha = 0.5)
plt.title('MAVERICK')
plt.show()
```
