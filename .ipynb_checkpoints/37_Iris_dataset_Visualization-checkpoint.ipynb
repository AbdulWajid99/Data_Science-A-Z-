```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataSet = pd.read_csv('E:\Data_Scientist\Skill_Share(DataScienceCourse)\Datasets\Iris_dataset_28.csv')
dataSet.head()
```

```python
#removing id 
dataSet.drop(columns='Id',axis=1)
```

```python
#plot all in diffrent plots
plot = dataSet[dataSet.Species=='Iris-setosa'].plot.scatter(x='SepalLengthCm',y='SepalWidthCm',color='r',label='Irsis Setosa')
dataSet[dataSet.Species=='Iris-versicolor'].plot.scatter(x='SepalLengthCm',y='SepalWidthCm',color='b',label='Irsis versicolor')
dataSet[dataSet.Species=='Iris-virginica'].plot.scatter(x='SepalLengthCm',y='SepalWidthCm',color='g',label='Irsis Verginica')
```

```python
#plot all in same plots ps use ax=plot
plot = dataSet[dataSet.Species=='Iris-setosa'].plot.scatter(x='SepalLengthCm',y='SepalWidthCm',color='r',label='Irsis Setosa')
dataSet[dataSet.Species=='Iris-versicolor'].plot.scatter(x='SepalLengthCm',y='SepalWidthCm',color='b',label='Irsis versicolor',ax=plot)
dataSet[dataSet.Species=='Iris-virginica'].plot.scatter(x='SepalLengthCm',y='SepalWidthCm',color='g',label='Irsis Verginica',ax=plot)
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.title('Sepal Length vs Sepal Width')
plt.show()
```

```python
#plot Petal Length vs Petal Width
plot = dataSet[dataSet.Species=='Iris-setosa'].plot.scatter(x='PetalLengthCm',y='PetalWidthCm',color='r',label='Irsis Setosa')
dataSet[dataSet.Species=='Iris-versicolor'].plot.scatter(x='PetalLengthCm',y='PetalWidthCm',color='b',label='Irsis versicolor',ax=plot)
dataSet[dataSet.Species=='Iris-virginica'].plot.scatter(x='PetalLengthCm',y='PetalWidthCm',color='g',label='Irsis Verginica',ax=plot)
plt.xlabel('Petal Length')
plt.ylabel('Petal Width')
plt.title('Petal Length vs Petal Width')
plt.show()
```
