```python
import matplotlib.pyplot as plt

import numpy as np
import math
```

```python
#normal plot
a = np.arange(0, math.pi*2, 0.04)
b =np.sin(a)
c=np.cos(a)
print(a)
print(b)
```

```python
#plot a and b
plt.plot(a,b)
plt.xlabel('Angle')
plt.ylabel('Sin Values')
plt.title('Sin Function')
plt.grid(True)
plt.show()
plt.plot(a,c)
plt.xlabel('Angle')
plt.ylabel('Sin Values')
plt.title('Sin Function')
plt.show()
```

```python
#Scatter Plot
x = 100 + np.random.randn(50)
y = 100 + np.random.randn(50)

plt.scatter(x,y)
plt.title("Scatter Plot")
plt.xlabel("X Values")
plt.ylabel("Y Values")
#plt.fill_between(x,y,96,where =(y>90),facecolor='b',alpha=0.4)
plt.show()
```

```python
#Bar Graph example number of pones sold in a month

c = ['jan','feb','march','april','may','jun']
v = [21,12,4,57,33,55]

plt.bar(c,v)
plt.title("Bar Graph")
plt.xlabel("Months")
plt.ylabel("Number of phones sold")
#plt.fill_between(x,y,96,where =(y>90),facecolor='b',alpha=0.4)
plt.show()
```

```python
#Pie chart spoerts played in pakistan

q = ['Cricket','Hockey','Football','Tennis','Basket ball','Jumangi']
r = [1500,250,3000,150,50,2200]

plt.pie(r,labels=q,autopct='%.1f%%')
plt.show()
```
