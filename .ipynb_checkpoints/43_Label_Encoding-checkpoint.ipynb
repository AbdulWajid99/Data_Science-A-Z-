```python
import pandas as pd
import numpy as np

irisDataset = pd.read_csv('E:\Data_Scientist\Skill_Share(DataScienceCourse)\Datasets\Iris_dataset_28.csv')

irisDataset.head()
```

```python
from sklearn.preprocessing import LabelEncoder

encode = LabelEncoder()
```

```python
#convert into integers the strings of species
#it will give integer to every category in species column
category = encode.fit_transform(irisDataset.Species)

#create new column for catogory
irisDataset['Label'] = category
print(irisDataset)
```

```python
# just run it once it will give all or specific number of rows instead od first few and last few
pd.set_option('display.max_rows',150)
print(irisDataset)
```
