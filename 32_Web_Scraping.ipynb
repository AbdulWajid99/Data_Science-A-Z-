```python
#pip install bs4

from bs4 import BeautifulSoup # for web scraping
import requests # to send http request
import pandas as pd
```

```python
siteUrl = 'https://en.wikipedia.org/wiki/Rice_production_in_India' #required page url

required_response = requests.get(siteUrl)

bsoup = BeautifulSoup(required_response.text, 'html.parser')

ricehtmltable = bsoup.find('table',{'class':'wikitable'})
```

```python
print(ricehtmltable)
```

```python
#pip install lxml #to convert html

riceTable = pd.read_html(str(ricehtmltable))
print(riceTable)
```

```python
type(riceTable)
```

```python
df = pd.DataFrame(riceTable[0]) #index zero means get all the data
df
```
