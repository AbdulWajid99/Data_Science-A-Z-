```python
import pandas as pd
```

```python
dataSet = pd.read_csv("https://raw.githubusercontent.com/fivethirtyeight/data/master/college-majors/recent-grads.csv")
dataSet.head()
```

```python
dataSet.shape
```

```python
#dataSet.dropna( will simply remove every null col)

dataSet.isnull().sum()
```

```python
# return any col with missing data
dataSet[dataSet.isnull().any(axis=1)]
```

# Method1: Boolean Mask

```python
#tell what condition u want if its true
#filter for one condiiton
df1 = dataSet[dataSet['Major_category']== 'Social Science']
df1
```

```python
#filter for more than one condiiton
#just use appranthsis not square brackets
df1M = dataSet[(dataSet['Major_category']== 'Social Science') 
               &
               (dataSet['ShareWomen']>=.5)]
df1M
```

```python
#use of ~ will produce exatly opposite of something or datframe
```

# Method2: Loc

```python
df2 = dataSet.loc[dataSet['Major_category']== 'Social Science']
df2
```

```python
df2M = dataSet.loc[(dataSet['Major_category']== 'Social Science') 
               &
               (dataSet['ShareWomen']>=.5)]
df2M
```

```python
#loc to get specfic column
df3 = dataSet.loc[(dataSet['Major_category']== 'Social Science') 
               &
               (dataSet['ShareWomen']>=.5),'Part_time']
df3.head()
```

```python
#loc to get specfic multiple column
col=['Part_time','Total']
df3m = dataSet.loc[(dataSet['Major_category']== 'Social Science') 
               &
               (dataSet['ShareWomen']>=.5),col]
df3m.head()
```

```python
#to get specific number of rows wih specific col name
dfS = dataSet.loc[dataSet.index[:10], ['Major','Major_category']]
dfS
```

```python
# to replace index with a col
dfI = dataSet.set_index(dataSet['Major'])
dfI
```

```python
rows = ['ZOOLOGY','LIBRARY SCIENCE']
col = ['Major_code','Total']

dfN = dfI.loc[rows,col]
dfN
```

# Method3: Query

```python
dfq = dataSet.query("Major_category == 'Social Science' & ShareWomen >=.5", inplace=False)
dfq.head()
```

```python
#how to check speed test?
#use mask  if all columnss u want

%timeit dataSet[(dataSet['Major_category']== 'Social Science') &(dataSet['ShareWomen']>=.5)]
#use loc if specific col u want
%timeit dataSet.loc[(dataSet['Major_category']== 'Social Science')&(dataSet['ShareWomen']>=.5)]
#dont use Query
%timeit dataSet.query("Major_category == 'Social Science' & ShareWomen >=.5")
```
